# 예제 설명  
영어 교과서에는 6개의 과로 나누어진 지문이 있다. 이 지문들을 문단별로 나누고, 주제 개수를 6으로 설정한 다음 LDA를 학습시켜보자!

```python
import warnings
warnings.filterwarnings(action='ignore')
```

## 1. 사전준비하기
데이터 불러오기, 데이터 전처리

```python
#필요한 모듈 import 하기
import pandas as pd
import gensim
import re
```

```python
#지문 불러오기
lesson1 = open('lesson1.txt', 'r', encoding = 'UTF-8')
lesson2 = open('lesson2.txt', 'r', encoding = 'UTF-8')
lesson3 = open('lesson3.txt', 'r', encoding = 'UTF-8')
lesson4 = open('lesson4.txt', 'r', encoding = 'UTF-8')
lesson5 = open('lesson5.txt', 'r', encoding = 'UTF-8')
lesson6 = open('lesson6.txt', 'r', encoding = 'UTF-8')
temp1, temp2, temp3, temp4, temp5, temp6 = [], [], [], [], [], []
while True:
    line = lesson1.readline()
    if not line: break
    temp1.append(line)
lesson1.close()
while True:
    line = lesson2.readline()
    if not line: break
    temp2.append(line)
lesson2.close()
while True:
    line = lesson3.readline()
    if not line: break
    temp3.append(line)
lesson3.close()
while True:
    line = lesson4.readline()
    if not line: break
    temp4.append(line)
lesson4.close()
while True:
    line = lesson5.readline()
    if not line: break
    temp5.append(line)
lesson5.close()
while True:
    line = lesson6.readline()
    if not line: break
    temp6.append(line)
lesson6.close()
temp1[0:10]
```

지문을 불러온 다음 줄 단위로 나눈다.

```python
#데이터 전처리
lessons = [temp1, temp2, temp3, temp4, temp5, temp6]
sentences = ['']

for temp in lessons:
    for sentence in temp:
        if "***" in sentence:
            sentences.append('')
            continue
        sentence = sentence.lower()
        sentence = re.sub(r'[^a-z]+',' ',sentence)
        if sentence == ' ': continue
        sentences[len(sentences)-1] += sentence
sentences[30] #30번째 문단
```

우선 지문 전체를 문단별로 자르고,  
대문자는 소문자로, 알파벳 이외의 글자(한국어 해석, 문장부호)는 공백으로 바꾼다.

## 2. LDA 학습을 위한 작업
불용어, 공백 제거  
dictionary와 corpus 만들기

```python
#단어별로 나누기
words = [sentences[i].split(' ') for i in range(len(sentences))]

import nltk
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
for i in range(len(sentences)):
    words[i] = [item for item in words[i] if (item not in stop_words) and (item != '')]

dictionary = gensim.corpora.Dictionary(words)
corpus = [dictionary.doc2bow(text) for text in words]
print(dictionary)
corpus[0][0:10]
```

nltk 라이브러리에서 제공하는 함수를 사용해 불용어를 제거한다.(관사, be동사 등등)

단어의 나열을 BOW(Bag Of Words)로 바꾼다.  
dictionary는 지문 속 단어의 사전이고(중복이 없는 나열)  
corpus는 dictionary를 이용해 문단을 단어의 id와 빈도로 나타낸 것이다.

## 3. LDA학습

```python
NUM_TOPICS = 6
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)

topics = ldamodel.print_topics(num_words=5)

for topic in topics:
    print(topic)
```

6개의 토픽에서 나타나는 단어들을 나열해 보여줌

```python
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
```
