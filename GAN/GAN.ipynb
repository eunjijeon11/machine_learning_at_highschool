## GAN의 원리  
GAN은 Generative Adversarial Network, 즉 적대적 대립 생성망의 약자임.  
2014년 Ian J. Goodfellow가 Generative Adversarial Nets 논문에서 최초로 아이디어와 원리를 제시함.  
이후 다양한 GAN 파생 모델들이 나오면서 딥페이크 같은 프로그램도 탄생.
![image.png](https://t1.daumcdn.net/cfile/tistory/99DD16395B75860434?download)  
GAN은 실제 이미지의 분포에 생성 모델의 확률분포를 맞춰가도록 학습됨   
GAN에서는 이름에서도 알 수 있듯이 적대적인 두 개의 모델이 경쟁하면서 학습이 이루어짐.  
<img src="https://t1.daumcdn.net/cfile/tistory/9928E6375B75872D17?download" width="800"/>  
<img src="https://t1.daumcdn.net/cfile/tistory/99A0C1335B758B401B?download" width="800"/>

- G: 노이즈 벡터(행렬)로부터 있을법한 이미지를 합성  
- D: 진짜 이미지 또는 가짜 이미지를 보고 진짜인지 가짜인지 판단(0~1사이의 값으로 나타냄)

이러한 원리로 학습을 진행하기 위해서 GAN 모델을 처음 제시한 이안 굿펠로우의 논문에서는 아래와 같은 함수에 대한 min max 게임을 제시했음.

$$\min(G)\max(D) V(D, G) = {\scriptstyle\text{Ex∼pdata(x)}}[log D(x)] + {\scriptstyle\text{Ez∼pz(z)}}[log(1 − D(G(z)))]$$  

- pdata는 진짜 데이터의 분포, pz는 노이즈의 분포를 나타냄   
- 첫번째 항: 진짜 데이터(x)를 뽑아서 log D(x)에 집어넣은 것의 기댓값  
즉, D가 얼마나 진짜 데이터를 진짜라고 판별하느냐를 의미
- 두번째 항: 노이즈 데이터(z)를 뽑아서 log(1-D(G(z)))에 집어넣은 것의 기댓값  
즉, D가 가짜 데이터를 얼마나 가짜라고 판별하느냐를 의미  
- V가 크면 D가 잘하고 있는 것이고 작으면 G가 잘하고 있는 것

위처럼 손글씨 이미지를 28x28의 array로 바꿔 사용

```python
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import layers
import numpy as np
import os
```

tensorflow, keras, numpy 라이브러리를 사용

### 1. MNIST 데이터 다운로드

```python
(train_images, train_labels), (_, _) = tf.keras.datasets.mnist.load_data()
train_images = train_images.reshape(train_images.shape[0], 28, 28, 1).astype('float32')
train_images = (train_images - 127.5) / 127.5 # 이미지를 [-1, 1]로 정규화

batch_size = 256

train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(train_images.shape[0]).batch(batch_size)
```

```python
len(train_dataset)
```

MNIST 데이터셋을 불러오고 정규화, 배치 사이즈에 맞게 나눔

### 2. 생성, 판별 모델 정의

```python
def discriminator_model():
     #판별모델 정의
    D = Sequential()
    D.add(layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same',input_shape=[28, 28, 1]))
    D.add(layers.LeakyReLU())
    D.add(layers.Dropout(0.3))
    D.add(layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'))
    D.add(layers.LeakyReLU())
    D.add(layers.Dropout(0.3))
    D.add(layers.Flatten())
    D.add(layers.Dense(1))
        
    return D
```

판별모델은 우리가 흔히 알고있는 인공신경망 모형  
일련의 레이어들을 통해 0~1 사이의 값을 도출  
(레이어 구성은 tensorflow 공식 사이트의 GAN 구현을 참고함)

```python
def generator_model():
    #분별모델 정의
    G = Sequential()
    G.add(layers.Dense(7*7*256, use_bias=False, input_shape=(100,)))
    G.add(layers.BatchNormalization())
    G.add(layers.LeakyReLU())
    G.add(layers.Reshape((7, 7, 256)))
    assert G.output_shape == (None, 7, 7, 256) # 주목: 배치사이즈로 None이 주어집니다.

    G.add(layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False))
    assert G.output_shape == (None, 7, 7, 128)
    G.add(layers.BatchNormalization())
    G.add(layers.LeakyReLU())

    G.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    assert G.output_shape == (None, 14, 14, 64)
    G.add(layers.BatchNormalization())
    G.add(layers.LeakyReLU())

    G.add(layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert G.output_shape == (None, 28, 28, 1)
        
    return G
```

생성모델또한 우리가 흔히 알고있는 인공신경망 모형  
노이즈 데이터(z)를 받아서 이미지 데이터 생성

```python
%matplotlib inline
import matplotlib.pyplot as plt


D = discriminator_model()
G = generator_model()

#가짜 이미지 생성
noise = tf.random.normal([1, 100])
generated_image = G(noise, training=False)

plt.imshow(generated_image[0, :, :, 0], cmap='gray')

#가짜 이미지에 대한 판별값
decision = D(generated_image)
print(decision)
```

### 3. 손실함수 정의  
손실함수: 실제 데이터와 생성된 데이터/판별 사이의 차이를 함수로 나타낸것  
손실함수가 최소값을 가지는 지점을 찾는 것이 학습의 목표이다!

```python
cross_entropy = keras.losses.BinaryCrossentropy(from_logits=True)
```

```python
def D_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss

def G_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)
```

**cross entropy**  
$$H(y,\widehat{y}) = -\frac{1}{N} \sum_{i=1}^N y_ilog\widehat{y_i} + (1-y_i)log(1-\widehat{y_i})$$   
실제 구현 시 loss function은 위와 같은 binary entropy loss function을 사용함  
D: D(x)와 1 사이의 오차를 계산=> y=D(x), ŷ=1 대입,  D(G(z))와 0 사이의 오차를 계산=> y=D(G(z)), ŷ=0 대입  
G: D(G(z))와 1 사이의 오차를 계산=> y=D(G(z)), ŷ=1대입    
$$J^{(D)}(\theta^{(D)}, \theta^{(G)}) = -\frac{1}{N} \sum_{i=1}^N [logD(x^{(i)}) + log(1-D(G(z^{(i)})]$$

**Discriminator**  
$$\frac{1}{m}\sum_{i=1}^m [\log D(x_i) + \log(1 - D(G(z_i)))]$$

- loss 함수가 커지는 쪽(기울기 쪽)으로 움직임(ascend)

**Generator**  
$$\frac{1}{m}\sum_{i=1}^m \log D(G(z_i)))$$

- log(1-D(G(z)))를 작게 하는 대신에 log(D(G(z)))를 크게 하는 식으로 변형   

- 가짜 데이터가 진짜라고 이해되는 정도임  
- 따라서 G는 이 식을 ascend 함(기울기 쪽으로 움직임) 

```python
#define optimizer
generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)
```

optimizer로는 Adam을 사용  
Adam은 경사하강법에 여러 기술들을 덧붙여 효율적으로 만든 최적화방법임!

### 4. 실제로 학습시키기

```python
epochs = 50
z_input_dim = 100

seed = tf.random.normal([16, z_input_dim]) #16개의 고정된 노이즈(시간대별로 예시 이미지 생성할 때 쓰일 것임)
```

```python
def show_imgs(model, epoch, test_input):
    predictions = model(test_input, training=False)
    
    fig = plt.figure(figsize=(4,4))
    
    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i+1)
        plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
        plt.axis('off')
        plt.savefig('image_at_epoch_{:04d}.png'.format(epoch))
        
    plt.show()
```

```python
import time

start = time.time()
losses = [[],[]]
d_values = [[],[]]

#학습
for epoch in range(epochs):
    
    #반복 10회마다 이미지 출력
    if (epoch) % 10 == 0:
        show_imgs(G, epoch, seed)
        print("걸린 시간", time.time()-start)
    
    for batch in train_dataset:
        noise = tf.random.normal([256, z_input_dim])
    
        with  tf.GradientTape() as g_tape, tf.GradientTape() as d_tape:
            generated_images = G(noise, training=True) #가짜 이미지 생성
    
            real_output = D(batch, training=True) #진짜 이미지에 대한 판별결과
            fake_output = D(generated_images, training=True) #가짜 이미지에 대한 판별결과
            d_values[0].append(real_output)
            d_values[1].append(fake_output)
    
            #loss 계산
            g_loss = G_loss(fake_output)
            d_loss = D_loss(real_output, fake_output)
            losses[0].append(g_loss)
            losses[1].append(d_loss)
    
        #기울기 계산
        g_grad = g_tape.gradient(g_loss, G.trainable_variables)
        d_grad = d_tape.gradient(d_loss, D.trainable_variables)
    
        generator_optimizer.apply_gradients(zip(g_grad, G.trainable_variables))
        discriminator_optimizer.apply_gradients(zip(d_grad, D.trainable_variables))
        
end = time.time()
print("걸린 시간",end-start)
```

## 결과

```python
show_imgs(G, 50, seed)
```

```python
g_loss = [losses[0][i] for i in range(len(losses[0])) if i%100==0]
d_loss = [losses[1][i] for i in range(len(losses[1])) if i%100==0]
```

```python
#loss 확인
figure, loss = plt.subplots(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
loss.plot(d_loss, 'r', label='discriminator_loss')
loss.plot(g_loss, 'b', label='generator_loss')

loss.set_title('D,G loss')
loss.set_xlabel('repeated time(every 100 times)')
loss.set_ylabel('loss')

loss.legend(loc='upper left')
plt.show()
```

## 느낀점  
확률과 통계에서 통계부분을 공부하고 있었을 때 우연히 딥페이크(인물 이미지 합성 프로그램)와 딥페이크로 인한 이슈들에 대해 알게되면서, 어떻게 이렇게까지 정교한 이미지를 만들 수 있을까 궁금해졌습니다. 딥페이크 모델인 GAN의 원리에 지금까지 배웠던 확률밀도함수, 기댓값, 모집단 등의 통계내용이 많이 등장했고, 인공지능 분야를 선도하는 기술에 내가 아는 내용을 적용해서 이해할 수 있다는 게 신기했습니다. 맨 처음 GAN을 제시했던 이안 굿펠로우의 논문도 읽어보고, 관련 강의도 찾아보고, 예제 코드에서 모르는 부분은 텐서플로우와 케라스 공식문서를 참고해가며 이해하려고 노력했습니다. 그 결과 생성자와 판별자의 인공신경망을 짜는 부분은 이해가 어려웠지만, 나머지 부분은 한줄한줄 따라해보며 이해할 수 있었습니다. 하지만 벡터나 행렬에 대한 깊은 이해가 없는 채로 여러 예제를 섞어 만들려다보니 학습이 잘 되지 않아 어려움을 겪었습니다. 텐서플로우 사이트의 dcGAN 예제를 따라하는 것으로 문제는 해결했지만, 인공지능 예제들을 더 잘 이해하기 위해서 백터와 행렬에 대해 더 알아보고싶다고 생각했습니다!
