GANÏóêÏÑúÎäî Ïù¥Î¶ÑÏóêÏÑúÎèÑ Ïïå Ïàò ÏûàÎìØÏù¥ Ï†ÅÎåÄÏ†ÅÏù∏ Îëê Í∞úÏùò Î™®Îç∏Ïù¥ Í≤ΩÏüÅÌïòÎ©¥ÏÑú ÌïôÏäµÏù¥ Ïù¥Î£®Ïñ¥Ïßê.

- G: ÎÖ∏Ïù¥Ï¶à Î≤°ÌÑ∞(ÌñâÎ†¨)Î°úÎ∂ÄÌÑ∞ ÏûàÏùÑÎ≤ïÌïú Ïù¥ÎØ∏ÏßÄÎ•º Ìï©ÏÑ±  
- D: ÏßÑÏßú Ïù¥ÎØ∏ÏßÄ ÎòêÎäî Í∞ÄÏßú Ïù¥ÎØ∏ÏßÄÎ•º Î≥¥Í≥† ÏßÑÏßúÏù∏ÏßÄ Í∞ÄÏßúÏù∏ÏßÄ ÌåêÎã®(0~1ÏÇ¨Ïù¥Ïùò Í∞íÏúºÎ°ú ÎÇòÌÉÄÎÉÑ)

Ïù¥Îü¨Ìïú ÏõêÎ¶¨Î°ú ÌïôÏäµÏùÑ ÏßÑÌñâÌïòÍ∏∞ ÏúÑÌï¥ÏÑú GAN Î™®Îç∏ÏùÑ Ï≤òÏùå Ï†úÏãúÌïú Ïù¥Ïïà ÍµøÌé†Î°úÏö∞Ïùò ÎÖºÎ¨∏ÏóêÏÑúÎäî ÏïÑÎûòÏôÄ Í∞ôÏùÄ Ìï®ÏàòÏóê ÎåÄÌïú min max Í≤åÏûÑÏùÑ Ï†úÏãúÌñàÏùå.

$$\min(G)\max(D) V(D, G) = {\scriptstyle\text{Ex‚àºpdata(x)}}[log D(x)] + {\scriptstyle\text{Ez‚àºpz(z)}}[log(1 ‚àí D(G(z)))]$$  

- pdataÎäî ÏßÑÏßú Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨, pzÎäî ÎÖ∏Ïù¥Ï¶àÏùò Î∂ÑÌè¨Î•º ÎÇòÌÉÄÎÉÑ   
- Ï≤´Î≤àÏß∏ Ìï≠: ÏßÑÏßú Îç∞Ïù¥ÌÑ∞(x)Î•º ÎΩëÏïÑÏÑú log D(x)Ïóê ÏßëÏñ¥ÎÑ£ÏùÄ Í≤ÉÏùò Í∏∞ÎåìÍ∞í  
Ï¶â, DÍ∞Ä ÏñºÎßàÎÇò ÏßÑÏßú Îç∞Ïù¥ÌÑ∞Î•º ÏßÑÏßúÎùºÍ≥† ÌåêÎ≥ÑÌïòÎäêÎÉêÎ•º ÏùòÎØ∏
- ÎëêÎ≤àÏß∏ Ìï≠: ÎÖ∏Ïù¥Ï¶à Îç∞Ïù¥ÌÑ∞(z)Î•º ÎΩëÏïÑÏÑú log(1-D(G(z)))Ïóê ÏßëÏñ¥ÎÑ£ÏùÄ Í≤ÉÏùò Í∏∞ÎåìÍ∞í  
Ï¶â, DÍ∞Ä Í∞ÄÏßú Îç∞Ïù¥ÌÑ∞Î•º ÏñºÎßàÎÇò Í∞ÄÏßúÎùºÍ≥† ÌåêÎ≥ÑÌïòÎäêÎÉêÎ•º ÏùòÎØ∏  
- VÍ∞Ä ÌÅ¨Î©¥ DÍ∞Ä ÏûòÌïòÍ≥† ÏûàÎäî Í≤ÉÏù¥Í≥† ÏûëÏúºÎ©¥ GÍ∞Ä ÏûòÌïòÍ≥† ÏûàÎäî Í≤É

Ïù¥ ÏΩîÎìúÏóêÏÑúÎäî ÏÜêÍ∏ÄÏî® Îç∞Ïù¥ÌÑ∞ÏÖãÏù∏ MNIST Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ GANÏùÑ ÌïôÏäµ

```python
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import layers
import numpy as np
import os
```

tensorflow, keras, numpy ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º ÏÇ¨Ïö©

### 1. MNIST Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥Î°úÎìú

```python
(train_images, train_labels), (_, _) = tf.keras.datasets.mnist.load_data()
train_images = train_images.reshape(train_images.shape[0], 28, 28, 1).astype('float32')
train_images = (train_images - 127.5) / 127.5 # Ïù¥ÎØ∏ÏßÄÎ•º [-1, 1]Î°ú Ï†ïÍ∑úÌôî

batch_size = 256

train_dataset = tf.data.Dataset.from_tensor_slices(train_images).shuffle(train_images.shape[0]).batch(batch_size)
```

```python
len(train_dataset)
```

MNIST Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Î∂àÎü¨Ïò§Í≥† Ï†ïÍ∑úÌôî, Î∞∞Ïπò ÏÇ¨Ïù¥Ï¶àÏóê ÎßûÍ≤å ÎÇòÎàî

### 2. ÏÉùÏÑ±, ÌåêÎ≥Ñ Î™®Îç∏ Ï†ïÏùò

```python
def discriminator_model():
     #ÌåêÎ≥ÑÎ™®Îç∏ Ï†ïÏùò
    D = Sequential()
    D.add(layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same',input_shape=[28, 28, 1]))
    D.add(layers.LeakyReLU())
    D.add(layers.Dropout(0.3))
    D.add(layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'))
    D.add(layers.LeakyReLU())
    D.add(layers.Dropout(0.3))
    D.add(layers.Flatten())
    D.add(layers.Dense(1))
        
    return D
```

ÌåêÎ≥ÑÎ™®Îç∏ÏùÄ Ïö∞Î¶¨Í∞Ä ÌùîÌûà ÏïåÍ≥†ÏûàÎäî Ïù∏Í≥µÏã†Í≤ΩÎßù Î™®Ìòï  
ÏùºÎ†®Ïùò Î†àÏù¥Ïñ¥Îì§ÏùÑ ÌÜµÌï¥ 0~1 ÏÇ¨Ïù¥Ïùò Í∞íÏùÑ ÎèÑÏ∂ú  
(Î†àÏù¥Ïñ¥ Íµ¨ÏÑ±ÏùÄ tensorflow Í≥µÏãù ÏÇ¨Ïù¥Ìä∏Ïùò GAN Íµ¨ÌòÑÏùÑ Ï∞∏Í≥†Ìï®)

```python
def generator_model():
    #Î∂ÑÎ≥ÑÎ™®Îç∏ Ï†ïÏùò
    G = Sequential()
    G.add(layers.Dense(7*7*256, use_bias=False, input_shape=(100,)))
    G.add(layers.BatchNormalization())
    G.add(layers.LeakyReLU())
    G.add(layers.Reshape((7, 7, 256)))
    assert G.output_shape == (None, 7, 7, 256) # Ï£ºÎ™©: Î∞∞ÏπòÏÇ¨Ïù¥Ï¶àÎ°ú NoneÏù¥ Ï£ºÏñ¥ÏßëÎãàÎã§.

    G.add(layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False))
    assert G.output_shape == (None, 7, 7, 128)
    G.add(layers.BatchNormalization())
    G.add(layers.LeakyReLU())

    G.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    assert G.output_shape == (None, 14, 14, 64)
    G.add(layers.BatchNormalization())
    G.add(layers.LeakyReLU())

    G.add(layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert G.output_shape == (None, 28, 28, 1)
        
    return G
```

ÏÉùÏÑ±Î™®Îç∏ÎòêÌïú Ïö∞Î¶¨Í∞Ä ÌùîÌûà ÏïåÍ≥†ÏûàÎäî Ïù∏Í≥µÏã†Í≤ΩÎßù Î™®Ìòï  
ÎÖ∏Ïù¥Ï¶à Îç∞Ïù¥ÌÑ∞(z)Î•º Î∞õÏïÑÏÑú Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±

```python
%matplotlib inline
import matplotlib.pyplot as plt


D = discriminator_model()
G = generator_model()

#Í∞ÄÏßú Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±
noise = tf.random.normal([1, 100])
generated_image = G(noise, training=False)

plt.imshow(generated_image[0, :, :, 0], cmap='gray')

#Í∞ÄÏßú Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌïú ÌåêÎ≥ÑÍ∞í
decision = D(generated_image)
print(decision)
```

### 3. ÏÜêÏã§Ìï®Ïàò Ï†ïÏùò  
ÏÜêÏã§Ìï®Ïàò: Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞ÏôÄ ÏÉùÏÑ±Îêú Îç∞Ïù¥ÌÑ∞/ÌåêÎ≥Ñ ÏÇ¨Ïù¥Ïùò Ï∞®Ïù¥Î•º Ìï®ÏàòÎ°ú ÎÇòÌÉÄÎÇ∏Í≤É  
ÏÜêÏã§Ìï®ÏàòÍ∞Ä ÏµúÏÜåÍ∞íÏùÑ Í∞ÄÏßÄÎäî ÏßÄÏ†êÏùÑ Ï∞æÎäî Í≤ÉÏù¥ ÌïôÏäµÏùò Î™©ÌëúÏù¥Îã§!

```python
cross_entropy = keras.losses.BinaryCrossentropy(from_logits=True)
```

```python
def D_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss

def G_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)
```

**cross entropy**  
$$H(y,\widehat{y}) = -\frac{1}{N} \sum_{i=1}^N y_ilog\widehat{y_i} + (1-y_i)log(1-\widehat{y_i})$$   
Ïã§Ï†ú Íµ¨ÌòÑ Ïãú loss functionÏùÄ ÏúÑÏôÄ Í∞ôÏùÄ binary entropy loss functionÏùÑ ÏÇ¨Ïö©Ìï®  
D: D(x)ÏôÄ 1 ÏÇ¨Ïù¥Ïùò Ïò§Ï∞®Î•º Í≥ÑÏÇ∞=> y=D(x), yÃÇ=1 ÎåÄÏûÖ,  D(G(z))ÏôÄ 0 ÏÇ¨Ïù¥Ïùò Ïò§Ï∞®Î•º Í≥ÑÏÇ∞=> y=D(G(z)), yÃÇ=0 ÎåÄÏûÖ  
G: D(G(z))ÏôÄ 1 ÏÇ¨Ïù¥Ïùò Ïò§Ï∞®Î•º Í≥ÑÏÇ∞=> y=D(G(z)), yÃÇ=1ÎåÄÏûÖ    
$$J^{(D)}(\theta^{(D)}, \theta^{(G)}) = -\frac{1}{N} \sum_{i=1}^N [logD(x^{(i)}) + log(1-D(G(z^{(i)})]$$

**Discriminator**  
$$\frac{1}{m}\sum_{i=1}^m [\log D(x_i) + \log(1 - D(G(z_i)))]$$

- loss Ìï®ÏàòÍ∞Ä Ïª§ÏßÄÎäî Ï™Ω(Í∏∞Ïö∏Í∏∞ Ï™Ω)ÏúºÎ°ú ÏõÄÏßÅÏûÑ(ascend)

**Generator**  
$$\frac{1}{m}\sum_{i=1}^m \log D(G(z_i)))$$

- log(1-D(G(z)))Î•º ÏûëÍ≤å ÌïòÎäî ÎåÄÏã†Ïóê log(D(G(z)))Î•º ÌÅ¨Í≤å ÌïòÎäî ÏãùÏúºÎ°ú Î≥ÄÌòï   

- Í∞ÄÏßú Îç∞Ïù¥ÌÑ∞Í∞Ä ÏßÑÏßúÎùºÍ≥† Ïù¥Ìï¥ÎêòÎäî Ï†ïÎèÑÏûÑ  
- Îî∞ÎùºÏÑú GÎäî Ïù¥ ÏãùÏùÑ ascend Ìï®(Í∏∞Ïö∏Í∏∞ Ï™ΩÏúºÎ°ú ÏõÄÏßÅÏûÑ) 

```python
#define optimizer
generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)
```

optimizerÎ°úÎäî AdamÏùÑ ÏÇ¨Ïö©  
AdamÏùÄ Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏóê Ïó¨Îü¨ Í∏∞Ïà†Îì§ÏùÑ ÎçßÎ∂ôÏó¨ Ìö®Ïú®Ï†ÅÏúºÎ°ú ÎßåÎì† ÏµúÏ†ÅÌôîÎ∞©Î≤ïÏûÑ!

### 4. Ïã§Ï†úÎ°ú ÌïôÏäµÏãúÌÇ§Í∏∞

```python
epochs = 50
z_input_dim = 100

seed = tf.random.normal([16, z_input_dim]) #16Í∞úÏùò Í≥†Ï†ïÎêú ÎÖ∏Ïù¥Ï¶à(ÏãúÍ∞ÑÎåÄÎ≥ÑÎ°ú ÏòàÏãú Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±Ìï† Îïå Ïì∞Ïùº Í≤ÉÏûÑ)
```

```python
def show_imgs(model, epoch, test_input):
    predictions = model(test_input, training=False)
    
    fig = plt.figure(figsize=(4,4))
    
    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i+1)
        plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
        plt.axis('off')
        plt.savefig('image_at_epoch_{:04d}.png'.format(epoch))
        
    plt.show()
```

```python
import time

start = time.time()
losses = [[],[]]
d_values = [[],[]]

#ÌïôÏäµ
for epoch in range(epochs):
    
    #Î∞òÎ≥µ 10ÌöåÎßàÎã§ Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†•
    if (epoch) % 10 == 0:
        show_imgs(G, epoch, seed)
        print("Í±∏Î¶∞ ÏãúÍ∞Ñ", time.time()-start)
    
    for batch in train_dataset:
        noise = tf.random.normal([256, z_input_dim])
    
        with  tf.GradientTape() as g_tape, tf.GradientTape() as d_tape:
            generated_images = G(noise, training=True) #Í∞ÄÏßú Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±
    
            real_output = D(batch, training=True) #ÏßÑÏßú Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌïú ÌåêÎ≥ÑÍ≤∞Í≥º
            fake_output = D(generated_images, training=True) #Í∞ÄÏßú Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌïú ÌåêÎ≥ÑÍ≤∞Í≥º
            d_values[0].append(real_output)
            d_values[1].append(fake_output)
    
            #loss Í≥ÑÏÇ∞
            g_loss = G_loss(fake_output)
            d_loss = D_loss(real_output, fake_output)
            losses[0].append(g_loss)
            losses[1].append(d_loss)
    
        #Í∏∞Ïö∏Í∏∞ Í≥ÑÏÇ∞
        g_grad = g_tape.gradient(g_loss, G.trainable_variables)
        d_grad = d_tape.gradient(d_loss, D.trainable_variables)
    
        generator_optimizer.apply_gradients(zip(g_grad, G.trainable_variables))
        discriminator_optimizer.apply_gradients(zip(d_grad, D.trainable_variables))
        
end = time.time()
print("Í±∏Î¶∞ ÏãúÍ∞Ñ",end-start)
```

## Í≤∞Í≥º

```python
show_imgs(G, 50, seed)
```

```python
g_loss = [losses[0][i] for i in range(len(losses[0])) if i%100==0]
d_loss = [losses[1][i] for i in range(len(losses[1])) if i%100==0]
```

```python
#loss ÌôïÏù∏
figure, loss = plt.subplots(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
loss.plot(d_loss, 'r', label='discriminator_loss')
loss.plot(g_loss, 'b', label='generator_loss')

loss.set_title('D,G loss')
loss.set_xlabel('repeated time(every 100 times)')
loss.set_ylabel('loss')

loss.legend(loc='upper left')
plt.show()
```
